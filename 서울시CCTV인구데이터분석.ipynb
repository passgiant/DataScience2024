```python
data_result = pd.merge(pop_seoul,cctv_seoul,on='구별')
data_result.head()
```

```python
data_result['고령자비율'] = data_result['고령자']/data_result['인구수']*100
data_result.head()
```

```python
data_result['외국인비율'] = data_result['외국인']/data_result['인구수']*100
data_result.head()
```

```python
data_result.columns = ['구별', '인구', '한국인', '외국인', '고령자', 'CCTV대수', 'CCTV증가율', '고령자비율', '외국인비율']
data_result.columns
```

```python
import matplotlib.pyplot as plt

plt.figure()
data_result['CCTV대수'].plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
data_result.set_index('구별', inplace=True)
data_result.head()
```

```python
import matplotlib.pyplot as plt

plt.figure()
data_result['CCTV대수'].plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
data_result.sort_values(by='CCTV대수', ascending=False, inplace=True)
data_result['CCTV대수'].plot(kind='barh', grid=True, figsize=(10,10))
```

```python
data_result['CCTV대수'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))
```

```python
plt.figure(figsize=(6,6))
plt.scatter(data_result['인구'], data_result['CCTV대수'], s=50)
plt.xlabel('인구')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
fp1 = np.polyfit(data_result['인구'], data_result['CCTV대수'], 1)
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)

plt.figure(figsize=(6,6))
plt.scatter(data_result['인구'], data_result['CCTV대수'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('인구')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
data_result['오차'] = np.abs(data_result['CCTV대수'] - f1(data_result['인구']))
```

```python
data_result['오차'] = np.abs(data_result['CCTV대수'] - f1(data_result['인구']))
df_sort = data_result.sort_values(by='오차', ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(data_result['인구'], data_result['CCTV대수'],
            c=data_result['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(25):
    # text를 출력할 x좌표의 시작점, y좌표의 시작점, 찍을 내용
    plt.text(df_sort['인구'][n]*1.02, df_sort['CCTV대수'][n]*0.981,
             df_sort.index[n], fontsize=15)

temp = "Hello!!!"
plt.text(400000-25000, 2800, temp, fontsize=15)

plt.xlabel('인구수')
plt.ylabel('인구당비율')
plt.colorbar()
plt.grid()
plt.show()
```
